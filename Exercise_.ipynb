```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
sns.pairplot(data=df, hue='size',palette='rainbow')
```

```python
sns.countplot(data=df, x=df['sex'], hue=df['smoker'])
```

```python
sns.barplot(x=df['sex'], y=df['total_bill'])
```

```python
sns.barplot(x=df['sex'], y=df['tip'], hue=df['smoker'], estimator=np.sum)
```

```python
sns.violinplot(x='tip',y='day',data=df)
sns.swarmplot(x='tip',y='day',data=df)
```

```python
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
df
```

```python
totalcorr = df[['total_bill','tip','size']]
totalcorr.corr()
```

```python
sns.heatmap(totalcorr.corr(),cmap='YlGnBu',annot=True)
```

```python
sns.clustermap(totalcorr.corr(),cmap='YlGnBu',annot=True)
```

```python
flightmap = flights.pivot_table(index='month',columns='year',values='passengers')
flightmap
```

```python
sns.heatmap(flightmap)
```
